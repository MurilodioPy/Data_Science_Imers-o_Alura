# Carregando a tabela com nomes

# Carregando a tabela com a quantidade

```python
df_total_acoes = pd.read_excel(f"{sys.path[0]}/tableStocks.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

# Carregando a tabela com o segmento e idade

```python
df_chat = pd.read_excel(f"{sys.path[0]}/tableStocks.xlsx", sheet_name="table_chat")
df_chat
```

```python
df_principal
```

# Diminuindo a quantidade de colunas da tabela principal

```python
df_principal = df_principal[['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

# Alterando o nome das colunas da tabela principal

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

# Criando a coluna 'var_pct' referente ao valor percentual e a coluna 'valor_inicial' referente ao valor do início do pregão

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_pct'] + 1)
df_principal
```

# Juntando a tabela principal com a tabela que contém a quantidade teórica

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

# Alterando nome da coluna 'qtd_teorica'

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtd_teorica'}).copy()
df_principal
```

# Cálculando a variação em reais e criando uma nova coluna

```python
df_principal['variacao_rs'] = df_principal['qtd_teorica'] * (df_principal['valor_final'] - df_principal['valor_inicial'])
df_principal
```

# Criando a coluna 'resultado_dia' aplicando uma função a cada valor da coluna 'variacao_rs'

```python
df_principal['resultado_dia'] = df_principal['variacao_rs'].apply(lambda x : 'Subiu' if x > 0 else ('Desceu' if x < 0 else "Estável"))
df_principal
```

Juntando a tabela principal com a tabela que contém o nome, segmento e idade das empresas

```python
df_principal = df_principal.merge(df_chat, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

# Criando a coluna 'categoria_idade'

```python
df_principal['categoria_idade'] = df_principal['Idade'].apply(lambda x : 'Mais de 100 anos' if x > 100 else ('Menos de 50 anos' if x < 50 else "Entre 50 e 100 anos"))
df_principal
```

# Analises


# Cálculos com a coluna 'variacao_rs' usando a coluna 'resultado_dia'


```python
maior = df_principal['variacao_rs'].max()
menor = df_principal['variacao_rs'].min()
media = df_principal['variacao_rs'].mean()

media_subiu = df_principal[df_principal['resultado_dia'] == 'Subiu']['variacao_rs'].mean()
media_caiu = df_principal[df_principal['resultado_dia'] == 'Desceu']['variacao_rs'].mean()

print(f'Maior variação: {maior:,.2f}')
print(f'Menor variação: {menor:,.2f}')
print(f'Média variação: {media:,.2f}')
print(f'Média dos valores que subiram: {media_subiu:,.2f}')
print(f'Média dos valores que desceram: {media_caiu:,.2f}')

```

# Criando novo df apenas com as ações que subiram

```python
df_principal_subiu = df_principal[df_principal['resultado_dia'] == 'Subiu']
df_principal_subiu
```

# Cálculos com a coluna 'variacao_rs' agrupando pela coluna 'segmento'

```python
df_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_segmento
```

# Cálculos com a coluna 'variacao_rs' usando a coluna 'resultado_dia' com o groupby

```python
df_resultado = df_principal.groupby('resultado_dia')['variacao_rs'].sum().reset_index()
df_resultado
```

# Cálculos com a coluna 'variacao_rs' usando a coluna 'resultado_dia' sem o groupby

```python
soma_subiu = df_principal[df_principal['resultado_dia'] == 'Subiu']['variacao_rs'].sum()
soma_caiu = df_principal[df_principal['resultado_dia'] == 'Desceu']['variacao_rs'].sum()
soma_estavel = df_principal[df_principal['resultado_dia'] == 'Estável']['variacao_rs'].sum()

print(f'Soma se subiu: {soma_subiu:,.2f}')
print(f'Soma se desceu: {soma_caiu:,.2f}')
print(f'Soma se estável: {soma_estavel:,.2f}')
```

# Cálculos com a coluna 'variacao_rs' usando a coluna 'categoria_idade'

```python
mais_de_100 = df_principal[df_principal['categoria_idade'] == 'Mais de 100 anos']['variacao_rs'].sum()
entre_50_100 = df_principal[df_principal['categoria_idade'] == 'Entre 50 e 100 anos']['variacao_rs'].sum()
menos_de_50 = df_principal[df_principal['categoria_idade'] == 'Menos de 50 anos']['variacao_rs'].sum()

print(f'Soma se Entre 50 e 100 anos: {entre_50_100:,.2f}')
print(f'Soma se Menos de 50 anos: {menos_de_50:,.2f}')
print(f'Soma se Mais de 100 anos: {mais_de_100:,.2f}')
```
